```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty graph
G = nx.Graph()

# Add nodes (5 nodes in this case)
G.add_nodes_from([1, 2, 3, 4, 5])

# Add edges (connecting nodes)
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Define the layout (position of nodes)
pos = nx.spring_layout(G)

# Draw the graph
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='red', font_size=12)

# Show the plot
plt.axis('off')
plt.show()
```

```python
print('nodes', G.nodes())
print('degree', G.degree([1, 2, 3]))
print("Sparse adjacency matrix:", nx.adjacency_matrix(G))
```

```python
print('average degree connectivity', nx.average_degree_connectivity(G))
print('average clustering coeeficient:', nx.average_clustering(G))
print('raw closeness centrality: ', nx.closeness_centrality(G, u=3))
```
